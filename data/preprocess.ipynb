```python
import pandas as pd
import numpy as np
import random
```

```python
train_df = pd.read_csv("valid_dataset_v3_simple.csv")
test_df = pd.read_csv("test_dataset.csv")
train_df.head(15)
```

```python
train_df["love"] = train_df["emotions"].str.split(",").str[0]
train_df["joy"] = train_df["emotions"].str.split(",").str[1]
train_df["fright"] = train_df["emotions"].str.split(",").str[2]
train_df["anger"] = train_df["emotions"].str.split(",").str[3]
train_df["fear"] = train_df["emotions"].str.split(",").str[4]
train_df["sorrow"] = train_df["emotions"].str.split(",").str[5]
```

```python
train_df.dropna(subset=["emotions"], inplace=True)
```

```python
train_df[["love","joy", "fright", "anger","fear","sorrow"]] = train_df[["love","joy", "fright", "anger","fear","sorrow"]].astype(int)
```

```python
dt = pd.concat([train_df, test_df], ignore_index=True, axis=0, sort=False)
dt
```

```python
dt['script_id'] = dt["id"].str.split('_').str[0]
dt['scene_num'] = dt["id"].str.split('_').str[1]
dt['sentence_num'] = dt["id"].str.split('_').str[3]
dt.head(10)
```

```python
dt.describe()
```

```python
script_id = dt["script_id"].unique().tolist()
scene_num = dt["scene_num"].unique().tolist()
sentence_num = dt["sentence_num"].unique().tolist()
character = dt["character"].unique().tolist()
character.remove(np.NAN)
script_id_dict = {el:i for i, el in enumerate(script_id)}
scene_num_dict = {el:i for i, el in enumerate(scene_num)}
sentence_num_dict = {el:i for i, el in enumerate(sentence_num)}
character_dict = {np.NAN: 0}
character_dict.update({el:i+1 for i, el in enumerate(character)})
```

```python
print("script_id num:", len(script_id_dict))
print("scene_num num:", len(scene_num_dict))
print("sentence_num num:", len(sentence_num_dict))
print("character num:", len(character_dict))
```

```python
dt["script_id_id"]  = dt["script_id"].apply(lambda x: script_id_dict[x])
dt["scene_num_id"]  = dt["scene_num"].apply(lambda x: scene_num_dict[x])
dt["sentence_num_id"]  = dt["sentence_num"].apply(lambda x: sentence_num_dict[x])
dt["character_id"]  = dt["character"].apply(lambda x: character_dict[x])
```

```python
train_cols = [
    "script_id_id", "scene_num_id", "sentence_num_id", "character_id", "content", "emotions", 
    "character_scene_num","character_script_num","character_sentence_num",
    "script_id_scene_num", "script_id_sentence_num",
]
```

```python
st = dt.pivot_table(index=["character"], values=["script_id", "scene_num", "sentence_num"],aggfunc = lambda x: len(x.unique()))
st.rename(
    columns={
        "script_id":"character_script_num", 
        "scene_num":"character_scene_num", 
        "sentence_num":"character_sentence_num"
    }, 
    inplace=True
)
st.reset_index(inplace=True)
dt = pd.merge(dt, st, on="character", how="left")
dt
```

```python
st2 = dt.pivot_table(index=["script_id"], values=["scene_num", "sentence_num"],aggfunc = lambda x: len(x.unique()))
st2.rename(
    columns={
        "scene_num":"script_id_scene_num", 
        "sentence_num":"script_id_sentence_num"
    }, 
    inplace=True
)
st2.reset_index(inplace=True)
dt = pd.merge(dt, st2, on="script_id", how="left")
dt
```

```python
dt = dt[train_cols]
```

```python
dt.head(10)
```

```python
train_dt = dt.loc[:train_df.shape[0]-1]
test_dt = dt.loc[train_df.shape[0]:]
del test_dt["emotions"]
```

```python
train_dt.shape
```

```python
train_dt.dropna(subset=["emotions"], inplace=True)
train_dt = train_dt.fillna(-1)
test_dt = test_dt.fillna(-1) 
train_dt
```

```python
train_dt.to_csv("train_dataset_v3.csv", index=False)
test_dt.to_csv("test_dataset_v3.csv", index=False)
```
